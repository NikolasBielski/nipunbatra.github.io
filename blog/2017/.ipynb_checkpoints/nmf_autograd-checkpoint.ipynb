## Non-negative matrix factorization using Autograd

```python
import autograd.numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
A = np.array([[3, 4, 5, 2],
                   [4, 4, 3, 3],
                   [5, 5, 4, 4]], dtype=np.float32).T
```

```python
A
```

```python
shape = A.shape
rank = 2
```

```python
def cost(W, H):
    pred = np.dot(W, H)
    return np.sqrt(((pred - A).flatten() ** 2).mean(axis=None))
```

```python
from autograd import grad, multigrad
```

```python
grad_cost= multigrad(cost, argnums=[0,1])
```

```python
H =  np.random.randn(rank, shape[1])
W =  np.random.randn(shape[0], rank)
learning_rate=0.01
for i in range(5000):
    if i%50==0:
        print "*"*20
        print i
        print "*"*20
        print cost(W, H)
    del_W, del_H = grad_cost(W, H)
    W = W-del_W*learning_rate
    H = H-del_H*learning_rate
    #print cost(W, H)
```
